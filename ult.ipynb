```python
import os
import cv2
```

```python
from utils.segmentation import segment_lung
```

```python
SRC_BASE = "lung_ct_split"
DEST_BASE = "lung_ct_lung_only"
```

```python
os.makedirs(DEST_BASE, exist_ok=True)
```

```python



splits = ["train", "val", "test"]

for split in splits:
    src_split = os.path.join(SRC_BASE, split)
    dest_split = os.path.join(DEST_BASE, split)

    for cls in os.listdir(src_split):
        src_cls = os.path.join(src_split, cls)
        dest_cls = os.path.join(dest_split, cls)

        os.makedirs(dest_cls, exist_ok=True)

        for img_name in os.listdir(src_cls):
            img_path = os.path.join(src_cls, img_name)
        
            lung = segment_lung(img_path)
        
            if lung is None or lung.size == 0:
                print("⚠️ Skipping:", img_path)
                continue
            
            save_path = os.path.join(dest_cls, img_name)
            cv2.imwrite(save_path, lung)

print("✅ Segmentation completed for train, val, test")
```
